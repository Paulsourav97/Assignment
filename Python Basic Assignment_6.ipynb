1. What are escape characters, and how do you use them?

Escape sequence is a combination of characters that allows you 
to include a special character in a string.

Most programming languages us a backslash(\) as an escape character.

For Example - Incase you want to print a (' - single quote) in a python statement.
You cannot do it. it will generally throw an error.

So to avoid this confusion for the compiler we use this escape sequence.

\' - single quote
\" - Double quote   etc
